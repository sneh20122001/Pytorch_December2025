## Matrix Multiplication

The main two rules for matrix multiplication to remember are:

1. The inner dimensions must match:
  - (3, 2) @ (3, 2) won't work
  - (2, 3) @ (3, 2) will work
  - (3, 2) @ (2, 3) will work

2. The resulting matrix has the shape of the outer dimensions:
  - (2, 3) @ (3, 2) -> (2, 2)
  - (3, 2) @ (2, 3) -> (3, 3)


### Note :
"@" in Python is the symbol for matrix multiplication.

```python
import torch
tensor = torch.tensor([1,2,3])
tensor.shape
```

```python
# Most Common Error in DL (fix with Transpose)

tensor_A = torch.tensor(
    [
        [1,2],
        [3,4],
        [5,6]
    ],
    dtype=torch.float32
)

tensor_B = torch.tensor(
    [
        [7,8],
        [9,10],
        [11,12]
    ],
    dtype=torch.float32
)

torch.matmul(tensor_A, tensor_B)
```

```python
print(tensor_A)
print(tensor_B.T)
```

```python
print(f"Shape of Tensors : \ntensor_A = {tensor_A},\ntensor_B = {tensor_B}\n")
print(f"New shapes : tensor_A = {tensor_A.shape}, tensor_B.T = {tensor_B.T.shape}\n")

A = torch.matmul(tensor_A,tensor_B.T)
print(A)
print(A.shape)
```

```python
torch.mm(tensor_A,tensor_B.T)
```

#### Note: A matrix multiplication like this is also referred to as the dot product of two matrices.

Neural networks are full of matrix multiplications and dot products.

## Reshaping, Stacking, Squeezing and Unsqueezing


image.png

```python
import torch
x = torch.arange(1.,8.)
x, x.shape
```

```python
# Add an Extra Dimension
x_reshaped = x.reshape(1,7)
x_reshaped, x_reshaped.shape
```

```python
# Change view (data same but view change)
z = x.view(1,7)
z, z.shape
```

```python
z[:, 0] = 5
z,x
```

```python
x_stacked = torch.stack([x,x,x,x],dim=0)
x_stacked
```

```python
# Remove extra dimension
print(f"x_reshaped: {x_reshaped.shape}")

x_squeezed = x_reshaped.squeeze()
print(x_squeezed)
print(x_squeezed.shape)
```

```python
# Add a dimension

print(f"x_reshaped: {x_squeezed.shape}")

x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(x_unsqueezed)
print(x_unsqueezed.shape)
```
